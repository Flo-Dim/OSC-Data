# 3- Data Modeling

### Objective of our model: Identify distinct segments of customers based on their purchasing behavior 

We are going to use a clustering algorithm called **K-Means algorithm** to perform this task mainly due to its ease of implementation, scalability and easy interpretation

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
# Select features for clustering
X = df_clean_new[['Quantity', 'UnitPrice']]

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42,  n_init=10)
df_clean_new.loc[:,'Cluster'] = kmeans.fit_predict(X_scaled)
print(df_clean_new.head())
```

```python
cluster_analysis = df_clean_new.groupby('Cluster').agg({
    'Quantity': 'mean',
    'UnitPrice': 'mean',
    'CustomerID': 'count'
}).reset_index()
```

```python
print(cluster_analysis)
```

# 4- Interpretation and Recommendations 


### Interpretation of Clustering Results:

1. **Cluster 0:**
   - **Quantity:** The average quantity of items purchased per transaction is approximately 12.63.
   - **UnitPrice:** The average unit price per item is around $3.00.
   - **CustomerID:** There are 397,906 unique Customer IDs associated with this cluster.

   **Interpretation:**
   - **Customer Segment:** Cluster 0 likely represents a large group of regular customers who make moderate-sized purchases.
   - **Behavioral Insight:** Customers in this cluster tend to buy a moderate number of items at a slightly higher than average unit price, indicating a balanced spending pattern.

2. **Cluster 1:**
   - **Quantity:** This cluster has an exceptionally high average quantity of items purchased per transaction, at 77,605 items.
   - **UnitPrice:** The average unit price per item is low, around $1.56.
   - **CustomerID:** There are only 2 unique Customer IDs associated with this cluster.

   **Interpretation:**
   - **Customer Segment:** Cluster 1 appears to be an outlier or a very specific segment characterized by extremely high quantity purchases.
   - **Behavioral Insight:** The low unit price suggests these purchases might involve bulk buying or wholesale transactions by a small number of customers.

3. **Cluster 2:**
   - **Quantity:** Transactions in this cluster typically involve 1 item per transaction on average.
   - **UnitPrice:** The average unit price per item is significantly high, approximately $2,782.91.
   - **CustomerID:** There are 16 unique Customer IDs associated with this cluster.

   **Interpretation:**
   - **Customer Segment:** Cluster 2 represents a segment where transactions involve very expensive items.
   - **Behavioral Insight:** Customers in this cluster purchase high-value items, possibly luxury goods or specialized products with a premium price tag.

### Recommendations Based on Clustering Insights:

1. **Cluster 0 (Regular Customers):**
   - **Recommendation:** Focus on loyalty programs, personalized offers, and product bundles to maintain and increase purchase frequency among these regular customers.
 
2. **Cluster 1 (Bulk Buyers):**
   - **Recommendation:** Focus on inserting volume discounts and bulk pricing to maintain their interest

3. **Cluster 2 (High-End Purchases):**
   - **Recommendation:** Highlight premium products, luxury items, or exclusive collections to appeal to customers interested in high-value purchases.
   - **Customer Experience:** Ensure excellent customer service and exclusive offers that match the premium nature of their purchases.
